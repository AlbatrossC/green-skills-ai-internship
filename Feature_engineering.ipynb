```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}
```

```python
energy_df = pd.DataFrame(data)

print("Original Energyu Data With Missing Values:")
energy_df.head()
```

```python
energy_df.tail()
```

```python
energy_df.info()
energy_df.describe()
energy_df.shape
```

```python
# Handling Missing Values
cleaned_energy_df = energy_df.dropna(axis=0)  
cleaned_energy_df.head()
```

```python
ec_m = energy_df['Energy Consumption (MWh)'].mean()
ec_c = energy_df['Cost (Million $)'].mean()
print("\nMean Energy Consumption (MWh):", ec_m)
print("Mean Cost (Million $):", ec_c)
```

```python

energy_df.fillna(method='ffill', inplace=True)
energy_df.head()
```

```python
energy_df.fillna(method='bfill', inplace=True)
energy_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

energy_df[['Energy Consumption (MWh)', 'Cost (Million $)']] = (
    scaler.fit_transform(
        energy_df[['Energy Consumption (MWh)', 'Cost (Million $)']]
    )
)

energy_df.head()
```
