```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_csv("..//Dataset//CC GENERAL.csv")
```

```python
data.head(3)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data_model = data.drop(columns=['CUST_ID'])
```

```python
data_model['CREDIT_LIMIT'].fillna(data_model['CREDIT_LIMIT'].median(), inplace=True)
data_model['MINIMUM_PAYMENTS'].fillna(data_model['MINIMUM_PAYMENTS'].median(), inplace=True)
```

```python

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_model)
```

```python
plt.figure(figsize=(10, 6))
sch.dendrogram(sch.linkage(data_scaled, method='ward'))
plt.title('Dendrogram for CC General')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
hc = AgglomerativeClustering(
    n_clusters=7,
    linkage='ward'
)

data['cluster'] = hc.fit_predict(data_scaled)
```

```python
sil_score = silhouette_score(data_scaled, data['cluster'])
print(f"Silhouette Score: {sil_score}")
```
